```python
# Atividade 1 - Para a função y = a * sin(b*x + c) faça
# (a) gere um conjunto de dados incluindo ruído;
# (b) suavização tipo "cubic spline";
# (c) suavização tipo Savitz Golay;
# (d) interpolação polinomial;
# (e) ajuste de curva;
# (f) faça um gráfico único com todos os dados dos itens (a) até (e)
# (g) faça uma discussão dos resultados;
```

```python
import numpy as np

from scipy import interpolate
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.signal import savgol_filter


#Atividade 1 (a):

def seno(x, a, b, c):
    return a*np.sin(b*x+c)

x = np.linspace(5,12,10)
y = seno(x,1,1,1) + 0.9* np.random.randn(10)
#plt.plot(x, y, 'g' , marker='o',linestyle='dashed', label= "Função com Ruído")
#plt.legend(loc='best')
#plt.show()

#Atividade 1 (b):


x_new = np.linspace(min(x),max(x),50) 

y_smooth = UnivariateSpline(x, y)

#plt.plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed', label= "cubic spline")
#plt.legend(loc='best')
#plt.show()

#Atividade 1 (c):
from scipy.interpolate import interp1d

xs = np.linspace(x.min(),x.max(), 100)
itp = interp1d(x, y, kind = 'linear')
window_size, poly_order = 77,3
yy_sg = savgol_filter(itp(xs), window_size, poly_order)

#plt.plot(xx, yy_sg, 'k', label= "Suavização Savitz Golay")
#plt.legend(loc='best')
#plt.show()

#Atividade 1 (d):
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

xnew = np.linspace(5, 12,10)

f1 = interpolate.interp1d(x, y,kind = 'linear')
#plt.plot(xnew, f1(xnew),'r-', label= "Interpolação polinomial")
#plt.legend(loc='best')
#plt.show()

#Atividade 1 (e):
from scipy.optimize import curve_fit
xx = np.linspace(x.min(),x.max(), 1000)
fit, _ = curve_fit(seno, x, y)
yy_fit = seno(xx, *fit)
#plt.plot(xx, yy_fit,'y--', label= "Ajuste da curva")
#plt.legend(loc='best')
#plt.show()
#Graficos das letras de (a) - (e):
fig, axes = plt.subplots(1,5 , figsize=(20, 3))
axes[0].plot(x, y, 'g' , marker='o',linestyle='dashed')
axes[0].set_title("Função com Ruído")
axes[1].plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed')
axes[1].set_title("cubic spline")
axes[2].plot(xs, yy_sg, 'k')
axes[2].set_title("Suavização Savitz Golay")
axes[3].plot(xnew, f1(xnew),'r-')
axes[3].set_title("Interpolação polinomial")
axes[4].plot(xx, yy_fit,'y--')
axes[4].set_title("Ajuste da curva")


#Atividade 1 (f):

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, y, 'g' , marker='o',linestyle='dashed', label= "Função com Ruído")
ax.plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed', label= "cubic spline")
ax.plot(xs, yy_sg, 'k', label= "Suavização Savitz Golay")
ax.plot(xnew, f1(xnew),'r-', label= "Interpolação polinomial")
ax.plot(xx, yy_fit,'y--', label= "Ajuste da curva")
ax.set_title("Gráfico único")
plt.legend(loc='best')
plt.show()

#Atividade 1 (g):

#É notavel que a interpolação é a que mais se distancia da função do grafico, 
#e na suavização Savitz Golay quando usado os parametros corretos chega 
#a se aproximar de modo semelhante da curva da função dada 
```

```python
# Atividade 2 - Para uma função gaussiana faça
# (a) gere um conjunto de dados incluindo ruído;
# (b) suavização tipo "cubic spline";
# (c) suavização tipo Savitz Golay;
# (d) interpolação polinomial;
# (e) ajuste de curva;
# (f) faça um gráfico único com todos os dados dos itens (a) até (e)
# (g) faça uma discussão dos resultados;

import numpy as np

from scipy import interpolate
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.signal import savgol_filter


#Atividade 2 (a):

def gaussian(x, a, b, c):
    return a*np.exp(-np.power(x - b, 2)/(2*np.power(c, 2)))

x = np.linspace(-10,10,10)
y = gaussian(x, 8, -1, 3) + 0.9* np.random.randn(10)

#Atividade 2 (b):

x_new = np.linspace(min(x),max(x),50) 

y_smooth = UnivariateSpline(x, y)

#Atividade 2 (c):
from scipy.interpolate import interp1d

xs = np.linspace(x.min(),x.max(), 100)
itp = interp1d(x, y, kind = 'linear')
window_size, poly_order = 39,3
yy_sg = savgol_filter(itp(xs), window_size, poly_order)

#Atividade 2 (d):
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

xnew = np.linspace(-10, 10,30)

f1 = interpolate.interp1d(x, y,kind = 'linear')

#Atividade 2 (e):
from scipy.optimize import curve_fit
xx = np.linspace(x.min(),x.max(), 100)
fit, _ = curve_fit(gaussian, x, y)
yy_fit = gaussian(xx, *fit)

#Graficos das letras de (a) - (e):
fig, axes = plt.subplots(1,5 , figsize=(20, 3))
axes[0].plot(x, y, 'g' , marker='o',linestyle='dashed')
axes[0].set_title("Função com Ruído")
axes[1].plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed')
axes[1].set_title("cubic spline")
axes[2].plot(xs, yy_sg, 'k')
axes[2].set_title("Suavização Savitz Golay")
axes[3].plot(xnew, f1(xnew),'r-')
axes[3].set_title("Interpolação polinomial")
axes[4].plot(xx, yy_fit,'y--')
axes[4].set_title("Ajuste da curva")

#Atividade 2 (f)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, y, 'g' , marker='o',linestyle='dashed', label= "Função com Ruído")
ax.plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed', label= "cubic spline")
ax.plot(xs, yy_sg, 'k', label= "Suavização Savitz Golay")
ax.plot(xnew, f1(xnew),'r-', label= "Interpolação polinomial")
ax.plot(xx, yy_fit,'y--', label= "Ajuste da curva")
ax.set_title("Gráfico único")
plt.legend(loc='best')
plt.show()
#Atividade 2 (g):

#tanto o ajuste como as outras suavizações teve um resultado muito proximo do esperado, 
#por mais que no ajuste o resultado foi o que mais se aproximou do desejado,
#porem a suavização Savitz Golay também pode obter um resultado parecido quando
#utilizados os parametros melhores assim eles vão definir mais a curva da função pedida
```

```python
# Atividade 3 - Para uma função lorentziana faça
# (a) gere um conjunto de dados incluindo ruído;
# (b) suavização tipo "cubic spline";
# (c) suavização tipo Savitz Golay;
# (d) interpolação polinomial;
# (e) ajuste de curva;
# (f) faça um gráfico único com todos os dados dos itens (a) até (e)
# (g) faça uma discussão dos resultados;

import numpy as np

from scipy import interpolate
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.signal import savgol_filter


#Atividade 3 (a):

def lorentzian( x, a, wo, c):
    return a / ((x-wo)**2 + c**2)

x = np.linspace(-10,10,20)
y = lorentzian(x, 8, -1, 3) + 0.09* np.random.randn(20)

#Atividade 3 (b):

x_new = np.linspace(min(x),max(x),50) 

y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(0.152)

#Atividade 3 (c):
from scipy.interpolate import interp1d

xs = np.linspace(x.min(),x.max(), 100)
itp = interp1d(x, y, kind = 'linear')
window_size, poly_order = 39, 2
yy_sg = savgol_filter(itp(xs), window_size, poly_order)

#Atividade 2 (d):
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

xnew = np.linspace(-10, 10,30)

f1 = interpolate.interp1d(x, y,kind = 'linear')

#Atividade 3 (e):
from scipy.optimize import curve_fit
xx = np.linspace(x.min(),x.max(), 100)
fit, _ = curve_fit(lorentzian, x, y)
yy_fit = lorentzian(xx, *fit)

#Graficos das letras de (a) - (e):
fig, axes = plt.subplots(1,5 , figsize=(20, 3))
axes[0].plot(x, y, 'g' , marker='o',linestyle='dashed')
axes[0].set_title("Função com Ruído")
axes[1].plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed')
axes[1].set_title("cubic spline")
axes[2].plot(xs, yy_sg, 'k')
axes[2].set_title("Suavização Savitz Golay")
axes[3].plot(xnew, f1(xnew),'r-')
axes[3].set_title("Interpolação polinomial")
axes[4].plot(xx, yy_fit,'y--')
axes[4].set_title("Ajuste da curva")

#Atividade 3 (f):

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, y, 'g' , marker='o',linestyle='dashed', label= "Função com Ruído")
ax.plot(x_new, y_smooth(x_new), 'purple' ,marker='x',linestyle='dashed', label= "cubic spline")
ax.plot(xs, yy_sg, 'k', label= "Suavização Savitz Golay")
ax.plot(xnew, f1(xnew),'r-', label= "Interpolação polinomial")
ax.plot(xx, yy_fit,'y--', label= "Ajuste da curva")
ax.set_title("Gráfico único")
plt.legend(loc='best')
plt.show()

#Atividade 3 (g):

#como na questao anterior o ajute de curva ficou como o melhor resultado para definir a função dada, 
#por mais que na suavização cubic spline também teve um resultado bem proximo do ajuste de curva,
#já a suavização savitz golay talvez alterando os parametors também consiga resultados parecidos com os outros, 
#mas em termo de facilidade o ajuste de curva foi o mais eficiente e o menos trabalhoso
```
