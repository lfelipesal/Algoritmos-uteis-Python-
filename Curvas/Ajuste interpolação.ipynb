```python
# Atividade 1 - Ajuste a interpolação  de uma curva do tipo : y = a * sin(b*x + c) usando 
# (a) interpolação polinomial linear e cúbica;
# (b) interpolação tipo "cubic spline";
# Faça a discussão dos resultados
#(a):
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
def seno(x, a, b, c):
    return a*np.sin(b*x+c)

xxx = np.linspace(5,12,10)
yyy = seno(xxx,1,1,1) + 0.1 * np.random.randn(10)

f1 = interpolate.interp1d(xxx, yyy,kind = 'linear')

f2 = interpolate.interp1d(xxx, yyy, kind = 'cubic')

xnew = np.linspace(5, 12,30)

plt.plot(xxx, yyy, 'o', xnew, f1(xnew), '-', xnew, f2(xnew), '--')

plt.legend(['data', 'linear', 'cubic','nearest'], loc = 'best')

plt.show()

#Discussão: por mais que a linear tenha a aparencia da curva ela não é tão exata como a cúbica.
```

```python
#(b):
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
import numpy as np 
def seno(x, a, b, c):
    return a*np.sin(b*x+c)

xx = np.linspace(5,12,15)
yy = seno(xx,1,1,1)  + 0.1 * np.random.randn(15)

spl = UnivariateSpline(xx, yy)
xs = np.linspace(5, 12, 1000)
spl.set_smoothing_factor(0.01)
plt.plot(xx, yy, 'ro', ms = 5)
plt.plot(xs, spl(xs), 'g', lw = 3)
plt.show()
```

```python
#Discussão: O resultado que mais apresentou uma melhor aproximação foi a interpolação "cubic spline"
```

```python
## Atividade 2 - Ajuste a interpolação  de uma curva do tipo :  usando y = a/( (x - b)**2 + c )
# (a) interpolação polinomial linear e cúbica;
# (b) interpolação tipo "cubic spline"; 
# Faça a discussão dos resultados
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
def fun(x, a, b, c):
    return a/((x-b)**2+c)

xxx = np.linspace(-5,15,10)
yyy = fun(xxx,1,1,1) + 0.1 * np.random.randn(10)

f1 = interpolate.interp1d(xxx, yyy,kind = 'linear')

f2 = interpolate.interp1d(xxx, yyy, kind = 'cubic')

xnew = np.linspace(-5, 15,30)

plt.plot(xxx, yyy, 'o', xnew, f1(xnew), '-', xnew, f2(xnew), '--')

plt.legend(['data', 'linear', 'cubic','nearest'], loc = 'best')

plt.show()
#Discussão: Como no primeiro item a linear nesse caso é ainda menos precisa,
#quando comparada a cúbica
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
import numpy as np 
def fun(x, a, b, c):
    return a/((x-b)**2+c)

xx = np.linspace(-5,15,15)
yy = fun(xx,1,1,1)  + 0.1 * np.random.randn(15)

spl = UnivariateSpline(xx, yy)
xs = np.linspace(-5, 15, 1000)
spl.set_smoothing_factor(0.001)
plt.plot(xx, yy, 'ro', ms = 5)
plt.plot(xs, spl(xs), 'g', lw = 3)
plt.show()

#Discussão:A "cubic spline" com o ruido teve muita mudança na precisão, 
#com isso tanto a interloção linear e a cubica normal melhor.
```

```python
## Atividade 3 - Ajuste a interpolação  de uma curva do tipo :  usando y = a* sin(b*x)*sqrt(x+c)
# (a) interpolação polinomial linear e cúbica;
# (b) interpolação tipo "cubic spline"; 
# Faça a discussão dos resultados
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
def fun1(x, a, b, c):
    return a* np.sin(b*x)* np.sqrt(x+c)

xxx = np.linspace(0,15,10)
yyy = fun1(xxx,1,1,1) + 0.1 * np.random.randn(10)

f1 = interpolate.interp1d(xxx, yyy,kind = 'linear')

f2 = interpolate.interp1d(xxx, yyy, kind = 'cubic')

xnew = np.linspace(0, 15,30)

plt.plot(xxx, yyy, 'o', xnew, f1(xnew), '-', xnew, f2(xnew), '--')

plt.legend(['data', 'linear', 'cubic','nearest'], loc = 'best')
plt.show()
#Discussão: a interpolação linear não foi nada preciso quanto a cubica,
#ela teve cortes e deixando sua precisão a deseja, nota que em alguns pontos a linear não passa.
```
