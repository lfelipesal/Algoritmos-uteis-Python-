```python
# Exemplo 01
# Usando Numpy para calcular derivadas numéricas
import numpy as np 
import matplotlib.pyplot as plt
# Criando vetor com os pontos x
x = np.linspace(0,10,1000)
# Calculando o incremento de derivação
dx = x[1]-x[0]
# Definindo a função a ser derivada
y = x**2 + 1
# Calculando o gradiente/derivada
dydx = np.gradient(y, dx)
# Fazendo os gráficos da função e de sua derivada primeira
h = plt.plot(x, y/np.abs(y).max())
h1 = plt.plot(x,dydx/np.abs(dydx).max())
plt.show()
```

```python
import numpy as np
import numdifftools as nd
import matplotlib.pyplot as plt
from scipy import interpolate
from scipy.interpolate import UnivariateSpline
from scipy.signal import savgol_filter
#exemplo 1
x = np.linspace(0,10,1000)
def f(x):
    return x**2 + 1 + 0.01* np.random.randn(1000)
#fun1 = x**2 + 1 + 0.1* np.random.randn(1000)
for i in range(3):
    df = nd.Derivative(f, n=i)
    y = df(x)
    n= str(i)
    p= 'Derivada da:'
    texto = p + n + 'º'
    if i == 0:
        texto = 'Função'
    h = plt.plot(x, y/np.abs(y).max(), label = texto)
    plt.legend(loc = 'best')
plt.show()

for i in range(3):
    df = nd.Derivative(f, n=i)
    y = df(x)
    x_new = np.linspace(min(x),max(x),500)
    y_smooth = UnivariateSpline(x, y)
    y_smooth.set_smoothing_factor(1)
    n= str(i)
    p= 'Derivada suavizada da:'
    texto = p + n + 'º'
    if i == 0:
        texto = 'Função suavizada'
    h = plt.plot(x_new, y_smooth(x_new)/np.abs(y).max(), label = texto)
    plt.legend(loc = 'best')
plt.show()
```
