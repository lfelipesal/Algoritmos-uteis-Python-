```python
######################################################################################################
# ALUNO : Luiz Felipe Dos Santos Alves
# MATRÍCULA : 2018094852
# TURMA : TF
######################################################################################################
import matplotlib as mpl 
import matplotlib.pyplot as plt
import numpy as np
import sympy
```

```python
# 1) Faça um gráfico da função y = (1/x^2) * sen(3*x), assim como de suas derivadas de ordem um e dois. (2.0 pontos)
####################################################################################################################
x = np.linspace(-2,2,100)
y1 = (1/(x**2)) * (np.sin(3*x))
y2 = ((3*(np.cos(3*x)))/x**2)+((-2*(np.sin(3*x)))/x**3)
y3 = ((3*((3*x**2)*(np.sin(3*x))+(4*x*np.cos(3*x)-2*np.sin(3*x))))*(-1))/(x**4)
       
fig, ax = plt.subplots()
ax.plot(x, y1, color="blue", label="y(x)")
ax.plot(x, y2, color="red", label="y'(x)")
ax.plot(x, y3, color="green", label="y”(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

```python
# 2) Através de análise gráfica localize as raízes de x**3-x**2+1 = x+2.  
# Faça um gráfico mostrando, via um "inset", uma das raízes. (2.0 pontos)
####################################################################################################################
x1 = np.linspace(-2,2,100)

#y4= (x**3)-(x**2)+1
y5= x+2
fig = plt.figure(figsize=(8, 4))

# Função a ser calculada
def f(x):
    return (x**3)-(x**2)+1

# Função para criar os gráficos
def plot_and_format_axes(ax, x, f, fontsize):
    ax.plot(x, f(x), linewidth=2 , label="f(x^3-x^2+1)")
    ax.plot(x, y5, linewidth=2, label="f(x+2)")
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
    ax.set_xlabel(r"$x$", )
    ax.set_ylabel(r"$f(x)$", )
    ax.legend()
    


# Gráfico principal
ax = fig.add_axes([0.1, 0.15, 0.8, 0.8], facecolor="#f5f5f5")
x = np.linspace(-4, 14, 1000)
plot_and_format_axes(ax, x, f, 18)

# Gráfico do detalhe
x0, x1 = 1.5 , 2

ax.axvline(x0, ymax=0.3, color="grey", linestyle=":")
ax.axvline(x1, ymax=0.3, color="grey", linestyle=":")

ax_insert = fig.add_axes([0.5, 0.5, 0.38, 0.42], facecolor='#f5f5f1')
x = np.linspace(x0, x1, 1000)
plot_and_format_axes(ax_insert, x, f, 14)

```

```python
# 3) Faça gráficos 3D de z = sen(x)^2 * (3*y - 3)**2 do tipo "surface", "wireplot" e "contour". O gráfico deverá ser no intervalo
# -10 < x < 10 e -10 < y < 10. (2.0 pontos)
####################################################################################################################
import mpl_toolkits.mplot3d 

fig, axes = plt.subplots(1, 3, figsize=(14, 4), subplot_kw={'projection': '3d'})

def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel("$x$", fontsize=16)
    ax.set_ylabel("$y$", fontsize=16)
    ax.set_zlabel("$z$", fontsize=16)   
    
x = y = np.linspace(-10, 10, 74)
X, Y = np.meshgrid(x, y)


Z = ((np.sin(X))**2)*((3*Y-3))**2

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = axes[0].plot_surface(X, Y, Z, rstride=1, cstride=1,linewidth=0, antialiased=False, norm=norm, cmap=mpl.cm.Blues)

cb = fig.colorbar(p, ax=axes[0], shrink=0.6)
                  
title_and_labels(axes[0], "surface")
                  
p = axes[1].plot_wireframe(X, Y, Z, rstride=2, cstride=2,color="darkgrey")
                  
title_and_labels(axes[1], "wireframe")

cset = axes[2].contour(X, Y, Z, zdir='z', offset=0, norm=norm, cmap=mpl.cm.Blues)

cset = axes[2].contour(X, Y, Z, zdir='y', offset=3, norm=norm, cmap=mpl.cm.Blues)

title_and_labels(axes[2], "contour")
```

```python
# 4) Faça um gráfico tipo contour plot da função de Rosenbrock: f(x,y) = (a - x)**2 + b(y - x**2)**2 com a = 1 e b = 100. 
# O intervalo será de -3 < x < 3  e -3 < y < 3. (2.0 pontos)
###################################################################################################################
import matplotlib as mpl 
import matplotlib.pyplot as plt
import numpy as np

x = y = np.linspace(-3, 3, 150)
X, Y = np.meshgrid(x, y)
Z = ((1-X)**2) + 100*((Y-X**2)**2)
fig, ax = plt.subplots(figsize=(6, 5))
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm=norm, cmap=mpl.cm.bwr)
ax.axis('tight')
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$y$", fontsize=18)
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
cb = fig.colorbar(p, ax=ax)
cb.set_label(r"$z$", fontsize=18)
cb.set_ticks([-10000, -5000, 0, 5000, 10000])
```
