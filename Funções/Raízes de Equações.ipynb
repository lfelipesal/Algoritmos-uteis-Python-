```python
import numpy as np
import scipy.optimize as opt
import matplotlib as mpl 
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline



def bissecao(f,a,b,prec):
    m = (a+b)/2
    if abs(b - a) < prec: return m
    if f(m) == 0: return m
    if f(m)*f(a) < 0: return bissecao(f,a,m,prec)
    else: return bissecao(f,m,b,prec)
    
print('ATV 2 Letra a')
def fun(x):
    return 2*x - np.cos(x) + 0.05*(1/((x-1)**2+1))
x = np.linspace(0,2*np.pi,15)
y= fun(x)
x_new = np.linspace(min(x),max(x),500) 
y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(1)

r = bissecao(fun,0,np.pi,1e-6)
print('raíz bissecção: ',r)
r = opt.brentq(lambda x: fun(x),0, np.pi)
print('raíz não suavizada: ',r)
r = opt.brentq(lambda x_new: y_smooth(x_new),0, np.pi)
print('raíz suavizada: ',r)

print('Letra b')
def fun(x):
    return x**3 - 4*(x**2)- x +6 + 0.05*(1/((x-1)**2+1))
x = np.linspace(-3,6,15)
y= fun(x)
x_new = np.linspace(min(x),max(x),500) 
y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(1)
r1b = bissecao(fun,-2 , 0 ,1e-6)
r2b = bissecao(fun ,0 , 2 ,1e-6)
r3b = bissecao(fun, 3 , 5 ,1e-6)
print("raíz bissecção 1: ", r1b)
print("raíz bissecção 2: ", r2b)
print("raíz bissecção 3: ", r3b)
r1n = opt.brentq(lambda x: fun(x), -2, 0)
r2n= opt.brentq(lambda x: fun(x), 0 , 2)
r3n = opt.brentq(lambda x: fun(x), 3, 5)
print("raíz não suavizada 1: ", r1n)
print("raíz não suavizada 2: ", r2n)
print("raíz não suavizada 3: ", r3n)
r1s = opt.brentq(lambda x_new: y_smooth(x_new), -3, 0)
r2s = opt.brentq(lambda x_new: y_smooth(x_new), 0,3 )
r3s = opt.brentq(lambda x_new: y_smooth(x_new), 3, 5)

print("raíz suavizada 1: ", r1s)
print("raíz suavizada 2: ", r2s)
print("raíz suavizada 3: ", r3s)
print('Letra c')
def fun(x):
    return x**5 + 3*(x**4) - 9*(x**3) + 20*x -12 + 0.05*(1/((x-1)**2+1))
x = np.linspace(-6,2,15)
y= fun(x)
x_new = np.linspace(min(x),max(x),500) 
y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(1)
r1b = bissecao(fun,-5 , -4.5 ,1e-6)
r2b = bissecao(fun ,-1.7 , -1.4 ,1e-6)
r3b = bissecao(fun, 0.5 , 1 ,1e-6)
print("raíz bissecção 1: ", r1b)
print("raíz bissecção 2: ", r2b)
print("raíz bissecção 3: ", r3b)
r1n = opt.brentq(lambda x: fun(x), -5, -4.5)
r2n= opt.brentq(lambda x: fun(x), -1.7 , -1.4)
r3n = opt.brentq(lambda x: fun(x), 0.5, 1)
print("raíz não suavizada 1: ", r1n)
print("raíz não suavizada 2: ", r2n)
print("raíz não suavizada 3: ", r3n)
r1s = opt.brentq(lambda x_new: y_smooth(x_new), -5, -4.5)
r2s = opt.brentq(lambda x_new: y_smooth(x_new), -1.7,-1.4 )
r3s = opt.brentq(lambda x_new: y_smooth(x_new), 0.5, 1)
print("raíz suavizada 1: ", r1s)
print("raíz suavizada 2: ", r2s)
print("raíz suavizada 3: ", r3s)

print('Letra d')

def fun(x):
    return np.exp(2*x-1) +x + 0.05*(1/((x-1)**2+1))
x = np.linspace(-2,2,15)
y= fun(x)
x_new = np.linspace(min(x),max(x),500) 
y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(1)

r = bissecao(fun,-0.5,0,1e-6)
print('raíz bissecção: ',r)
r = opt.brentq(lambda x: fun(x),-0.5, 0)
print('raíz não suavizada: ',r)
r = opt.brentq(lambda x_new: y_smooth(x_new),-0.5, 0)
print('raíz suavizada: ',r)
print('Letra e')

def fun(x):
    return np.log(x)+x + 0.05*(1/((x-1)**2+1))
x = np.linspace(0.1,2,15)
y= fun(x)
x_new = np.linspace(min(x),max(x),500) 
y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(1)

r = bissecao(fun,0.4,0.9,1e-6)
print('raíz bissecção: ',r)
r = opt.brentq(lambda x: fun(x),0.4, 0.9)
print('raíz não suavizada: ',r)
r = opt.brentq(lambda x_new: y_smooth(x_new),0.4, 0.9)
print('raíz suavizada: ',r)

print('Letra f')

def fun(x):
    return np.tanh(x-1) - np.log(x) + 0.05*(1/((x-1)**2+1))
x = np.linspace(0.0001,4,15)
y= fun(x)
x_new = np.linspace(min(x),max(x),500) 
y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(1)
r1b = bissecao(fun,1 , 1.5 ,1e-6)
r2b = bissecao(fun ,2 , 3 ,1e-6)
print("raíz bissecção 1: ", r1b)
print("raíz bissecção 2: ", r2b)
r1n = opt.brentq(lambda x: np.tanh(x-1) - np.log(x), 1, 2)
r2n= opt.brentq(lambda x: fun(x), 2 , 3)
print("raíz não suavizada 1: ", r1n)
print("raíz não suavizada 2: ", r2n)
r1s = opt.brentq(lambda x_new: y_smooth(x_new), 1, 2)
r2s = opt.brentq(lambda x_new: y_smooth(x_new), 2,3 )
print("raíz suavizada 1: ", r1s)
print("raíz suavizada 2: ", r2s)
```

É notavel que na raiz da função suavizada teve uma certa diferença quando comparada a o metodo da bisseção e da curva não suavizada e essa diferença pode se dar pela modificação que ocorre quando a curva é suavizada.
