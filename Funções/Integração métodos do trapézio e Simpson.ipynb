```python
# Atividade 1 - Para a função y = 2.0*sin^2(3.0*x + 1.77)*cos(1.0*x), no intervalo de [0,2PI] faça:
# (a) gere um conjunto de dados incluindo ruído;
# (b) Faça uma suavização ou interpolação (com resultados em gráfico);
# (c) Calcule o valor da integral pelos métodos do trapézio e Simpson;
# (d) Compare com o valor exato da integral;
import numpy as np

from scipy import interpolate
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.signal import savgol_filter

def fun1(x):
    return 2.0*((np.sin(3.0*x + 1.77))**2)*np.cos(1.0*x)

# (a):

x = np.linspace(0,2*np.pi,100)
y = fun1(x) + 0.1* np.random.randn(100)

# (b):

x_new = np.linspace(min(x),max(x),500) 

y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(4.7)

plt.plot(x_new, y_smooth(x_new), 'pink',x , y , 'k.')
plt.legend(['Suavização', 'dados com ruído','nearest'], loc = 'best')
plt.show()

# (c):
from scipy.integrate import simps
from numpy import trapz



area = trapz(y_smooth(x_new),x_new)
print ('Trapézio - área = ',area)
area = simps(y_smooth(x_new),x_new)
print ("Simpson - área= ", area)

#(d):


# não achei uma integral que possa usar para o calculo da area exata
```

```python
# Atividade 2 - Para a função y = 1/[x^2 * sqrt(4 - x^2)];, no intervalo de [0,20] faça:
# (a) gere um conjunto de dados incluindo ruído;
# (b) Faça uma suavização ou interpolação (com resultados em gráfico);
# (c) Calcule o valor da integral pelos métodos do trapézio e Simpson;
# (d) Compare com o valor exato da integral;

def fun2(x):
    return 1.0/((x**2)*np.sqrt(4-(x**2)))

# (a):

x = np.arange(0.1,1.999,0.1)
y = fun2(x)+0.5*np.random.randn(19)

# (b):

x_new = np.linspace(min(x),max(x),200) 

y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(0.003)
                              
                              

plt.plot(x_new, y_smooth(x_new), 'pink',x , y , 'k.')
plt.legend(['Suavização', 'dados com ruído','nearest'], loc = 'best')
plt.show()

# (c):

area = trapz(y_smooth(x_new),x_new)
print ('Trapézio - área = ',area)
area = simps(y_smooth(x_new),x_new)
print ('Simpson - área = ',area)

#(d):
x = x[len(x)-1]
F1 = -(np.sqrt(4-x**2))/(4*x)
x = 0.1
F2 = -(np.sqrt(4-x**2))/(4*x)

exata = F1 - F2

print ("Integral exata - área= ", exata)
```

```python
# Atividade 3 - Para a função y = 1/sqrt(x^2 - 64), no intervalo de [0,20] faça:
# (a) gere um conjunto de dados incluindo ruído;
# (b) Faça uma suavização ou interpolação (com resultados em gráfico);
# (c) Calcule o valor da integral pelos métodos do trapézio e Simpson;
# (d) Compare com o valor exato da integral;

def fun2(x):
    return 1.0/(np.sqrt((x**2)-64))

# (a):

x = np.linspace(8.1,20,15)
y = fun2(x)+0.02*np.random.randn(15)

# (b):

x_new = np.linspace(min(x),max(x),200) 

y_smooth = UnivariateSpline(x, y)
y_smooth.set_smoothing_factor(0.01)
                              
                              

plt.plot(x_new, y_smooth(x_new), 'pink',x , y , 'k.')
plt.legend(['Suavização', 'dados com ruído','nearest'], loc = 'best')
plt.show()

# (c):

area = trapz(y_smooth(x_new),x_new)
print ('Trapézio - área = ',area)
area = simps(y_smooth(x_new),x_new)
print ('Simpson - área = ',area)

#(d): 

x = x[len(x)-1]
F1 = np.log((1/8)*(np.sqrt((x**2)-64))+x)
x = 8
F2 = np.log((1/8)*(np.sqrt((x**2)-64))+x)

exata = F1 - F2

print ("Integral exata - área= ", exata)
```
